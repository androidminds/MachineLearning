
# 简单线性回归

对于形如y = wx + b的一元二次方程可以用最小二乘法求解，公式如下：

$ w = \frac{\sum_{i=1}^{m}y_{i}(x_{i}-\bar{x})}{\sum_{i=1}^{m}x^{2}-\frac{1}{m}(\sum_{i=1}^{m}x_{i})^{2}} $

$ b = \frac{1}{m}\sum_{i=1}^{m}(y_{i} - wx_{i}) $

其中$\bar{x}$是x的平均值


```python
import numpy as np
import matplotlib.pyplot as plt
```

定义类SimpleLinearRegression，其中fit函数计算w和b的值，predict函数预测新的值。

```python
class SimpleLinearRegression:
    def fit(self, X, Y):
        m = X.shape[0]
        self.w =  np.multiply(Y,(X-X.mean())).sum() / (np.square(X).sum() - np.square(X.sum())/m)
        self.b = (Y- np.multiply(self.w,X)).sum() / m

    def predict(self, x):
        return self.w * x + self.b
```

测试和显示结果

```python
X = np.array([5, 8, 9, 10, 13, 16, 18, 20])
Y = np.array([0,1,2,3,4,5,6,7])

slr = SimpleLinearRegression()
slr.fit(X, Y)
print(f"w={slr.w}, b={slr.b}")

plt.title('simple linear regression')
plt.scatter(X, Y) 
plt.plot([0, 25],[slr.predict(0), slr.predict(25)])
plt.show()
```
